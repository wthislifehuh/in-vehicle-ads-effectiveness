## Plot the histograms for each categorical attribute in the dataset

```python
def plot_categorical_bargraphs(dataframe):
    # Get a list of all categorical columns
    categorical_columns = dataframe.select_dtypes(include=["object"]).columns

    # Loop through each categorical column and plot a bar graph
    for column in categorical_columns:
        # Calculate the frequency of each category
        category_counts = dataframe[column].value_counts()

        # Create a bar graph
        plt.figure(figsize=(12, 5))
        ax = category_counts.plot(kind='bar', color='skyblue')
        plt.title("Bar Graph for {}".format(column))
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.xticks(rotation=45)

        # Add text labels for exact values on top of each bar
        for p in ax.patches:
            ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

        plt.show()
```

## Standardization of numerical data

## Handle missing values in categorical data

Strategy: <br>

```python
X_train_cat_tr_df = pd.DataFrame(X_train_cat_tr, columns=X_train_cat.columns)
print("Missing values in all categorical features:")
print(X_train_cat_tr_df.isnull().sum())
```

#### Select categorical features that need to be encoded

```python
X_train_cat_tr_onehot = X_train_cat_tr_df.drop(["children", "moreThan5km", "moreThan15km", "sameDirection"], axis=1)
X_train_cat_tr_encoded = X_train_cat_tr_df[["children", "moreThan5km", "moreThan15km", "sameDirection"]]
```

```python
X_train_cat_tr_onehot.describe()
```

### Split categorical set into ordinal data and nominal data

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

X_train_cat_tr_onehot_ordinal = X_train_cat_tr_onehot.drop(["maritalStatus", "education", "income", "destination", "passanger", "weather", "storeType"], axis=1)
X_train_cat_tr_onehot_nominal = X_train_cat_tr_onehot[["maritalStatus", "education", "income", "destination", "passanger", "weather", "storeType"]]
```

### Encode the ordinal data using OrdinalEncoder and nominal data using OneHotEncoder

```python
# Create separate instances of OrdinalEncoder and OneHotEncoder
ordinal_encoder = OrdinalEncoder()
onehot_encoder = OneHotEncoder()

for col in X_train_cat_tr_onehot_ordinal:
    # Extract the column as a DataFrame and apply OrdinalEncoder to it
    col_data = X_train_cat_tr_onehot_ordinal[[col]]
    col_encoded = ordinal_encoder.fit_transform(col_data)

    # Assign the transformed column back to X_train_cat_tr_onehot_ordinal
    X_train_cat_tr_onehot_ordinal[col] = col_encoded

    print(f'Categories for column "{col}": {ordinal_encoder.categories_}')
    print(f'Shape of {col}: {col_encoded.shape}\n')

for col in X_train_cat_tr_onehot_nominal:
    # Extract the column as a DataFrame and apply OneHotEncoder to it
    col_data = X_train_cat_tr_onehot_nominal[[col]]
    col_encoded = onehot_encoder.fit_transform(col_data).toarray()  # Convert to dense array

    # Get feature names for one-hot encoded columns
    feature_names = onehot_encoder.get_feature_names_out([col])

    # Create a DataFrame for the one-hot encoded columns
    col_encoded_df = pd.DataFrame(col_encoded, columns=feature_names)

    # Assign the transformed column back to X_train_cat_tr_onehot_nominal
    X_train_cat_tr_onehot_nominal = pd.concat([X_train_cat_tr_onehot_nominal, col_encoded_df], axis=1)

    print(f'Categories for column "{col}": {onehot_encoder.categories_}')
    print(f'Shape of {col}: {col_encoded.shape}\n')

X_train_cat_tr_onehot_nominal = X_train_cat_tr_onehot_nominal.drop(["maritalStatus", "education", "income", "destination", "passanger", "weather", "storeType"], axis=1)
X_train_cat_tr_onehot = pd.concat([X_train_cat_tr_onehot_ordinal, X_train_cat_tr_onehot_nominal], axis=1)
```

### Concatenate all categorical features into a single dataframe

```python
X_train_cat_tr_df = pd.concat([X_train_cat_tr_encoded, X_train_cat_tr_onehot], axis=1)
X_train_cat_tr_df.head()
```

### Finalise categorical set

Convert the DataFrame object back to Numpy array object

```python
X_train_cat_tr = X_train_cat_tr_df.to_numpy()
print('X_train_cat_tr shape:', X_train_cat_tr.shape)
```
